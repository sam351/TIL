```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print(type(iris_dataset))
iris_dataset
```

```python
print(type(iris_dataset.data))
iris_dataset.data
```

## 1. Data Acquisition - split train, test ds

```python
from sklearn.model_selection import train_test_split

# train 0.75, test 0.25
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print('The sizes of :  X_train, \t X_test, \t y_train, \t y_test')
print(f'\t\t{X_train.shape}, \t {X_test.shape}, \t {y_train.shape}, \t {y_test.shape}')
```

## 2. EDA - plotting train ds

```python
import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.DataFrame(X_train, 
                       columns=[name.replace(' (cm)','').replace(' ','_') for name in iris_dataset.feature_names])
iris_df
```

#### - With scatter plot, Concluded that there are some features that are useful for classficiation

```python
pd.plotting.scatter_matrix(iris_df, figsize=(10,10), c = y_train, marker='o', s=40, alpha=.8, hist_kwds={'bins':20}, cmap=mglearn.cm3)
plt.show()
```

## 3. Modeling (k-nearest-neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

## 4. Validation (using test ds)

```python
knn.predict(X_test)
```

```python
iris_dataset['target_names'][ knn.predict(X_test) ]
```

```python
knn.score(X_test, y_test)
```

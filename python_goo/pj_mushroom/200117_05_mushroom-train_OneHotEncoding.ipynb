```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

# load data
mr = pd.read_csv('mushroom.csv', header=None)

# # # encode the values - OneHotEncoding
# label_series = pd.get_dummies(pd.Series([ val for val in mr.iloc[:,0] ]))['p']
# data_df = pd.get_dummies(mr.iloc[:,1:])

# the other way - dummy encoding (less # of cols)
label_series = pd.get_dummies(pd.Series([ val for val in mr.iloc[:,0] ]))['p']
data_df = pd.get_dummies(mr.iloc[:,1:], drop_first=True)

pd.concat([label_series, data_df], axis=1)
```

```python
# split the data
data_train, data_test, label_train, label_test = train_test_split(data_df, label_series)

# train model with train ds
clf = RandomForestClassifier(n_estimators=10)
clf.fit(data_train, label_train)

# validate the model with test ds
predict = clf.predict(data_test)
ac_score = metrics.accuracy_score(label_test, predict)
cl_report = metrics.classification_report(label_test, predict)

print('ac_score :', ac_score)
print('cl_report :\n', cl_report)
```

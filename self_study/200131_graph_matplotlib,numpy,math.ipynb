```python
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.preprocessing import MinMaxScaler
```

```python
x = np.linspace(-10, 10, 1000)

y1 = 700*x+3
y2 = 50*x**2+20*x+5
y3 = 10*x**3+5*x**2+4*x+3
y4 = x**4+2*x**3+5*x**2+4*x+3
y11 = x**11+x**4+x**3+15

scaler = MinMaxScaler()
y11_scaled = scaler.fit_transform(np.array(y11).reshape(-1,1)).reshape(1,-1).tolist()[0]


fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,10))
for tmp_x, tmp_y, ax in zip([x]*4, [y1, y2, y3, y4], axes.ravel()):
    ax.plot(tmp_x, tmp_y)
    ax.set_yticks(np.arange(-10000,11000, 1000))
    pass
axes[2,0].plot(x, y11)
axes[2,1].plot(x, y11_scaled)
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y_cos_sin = [ math.sin(math.cosh(x)+100) for x in x ]

plt.figure(figsize=(10,10))
plt.plot(x, y_cos_sin)
plt.show()
```

```python
x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(0, 10, 1000)
x3 = np.linspace(-5, 1, 1000)

y1 = 15*x1 + 30*x2 + 23*x3 + 10  # linear function
y2 = 1.5*x1**2 + 2*x2**2 + x1 + 0.5*x2 + 3  # quadratic function
y3 = 3*x3**3 + 1.5*x1**2 + 2*x2**2 + x1 + 0.5*x2 + 3  # 3rd degree
y4 = 10*x3**3 + 1.5*x1**2 + 2*x2**2 + x1 + 0.5*x2 + 3  # 3rd degree
y5 = x1**4 + 3*x3**3 + 1.5*x1**2 + 2*x2**2 + x1 + 0.5*x2 + 3  # 4th degree
y6 = x3**4 + 300*x3**3 + 1.5*x1**2 + 2*x2**2 + x1 + 0.5*x2 + 3  # 4th degree

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,10))
for tmp_y, ax in zip([y1, y2, y3, y4, y5, y6], axes.ravel()):
    ax.plot(tmp_y)
    ax.set_yticks(np.arange(-500,500, 100))
    pass
plt.show()
```

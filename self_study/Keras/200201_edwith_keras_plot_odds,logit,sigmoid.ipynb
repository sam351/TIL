### Plotting Graphs related with Logistic Regression 

```python
import numpy as np
import matplotlib.pyplot as plt
```

- plot odds given p (odds = p/(1-p))

```python
def odds(p):
    return p/(1-p)

p = np.arange(0.001, 1, 0.001)
z = odds(p)

plt.figure(figsize=(5,3))
plt.plot(p, z, 'b')
plt.ylim(0, 150)
plt.title('Odds given p')
plt.xlabel('p', fontsize=15)
plt.ylabel('odds', fontsize=15)
plt.show()
```

- logit function (log of odds)

```python
z = np.log(odds(p))

plt.figure(figsize=(6,4))
plt.plot(p, z, 'b')
plt.grid()
plt.title('Logit function')
plt.xlabel('p', fontsize=15)
plt.ylabel('logit', fontsize=15)
plt.show()
```

- sigmoid function (inverse function of logit function)

```python
def sigmoid(wx):
    return 1/(1 + np.exp(-wx))

z = np.arange(-8, 8, 0.1)
y = sigmoid(z)

plt.figure(figsize=(5,3))
plt.plot(z, y, 'b')
plt.grid()
plt.title('Logistic | Sigmoid function')
plt.xlabel('z (=wx)', fontsize=15)
plt.ylabel('sigmoid(z)', fontsize=15)
plt.show()
```

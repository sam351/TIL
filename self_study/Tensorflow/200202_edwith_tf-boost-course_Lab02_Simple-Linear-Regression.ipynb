## Full code
##### Build hypothesis and cost -> Gradient descent

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()  # make it run immediately

# Data
x_data = list(np.arange(1,6))  # [1~5]
y_data = list(np.arange(1,6))

# W, b initialize
W = tf.contrib.eager.Variable(2.9)
b = tf.contrib.eager.Variable(0.5)

learning_rate = 0.01

for i in range(100+1):  # W, b update
    # Gradient descent
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean( tf.square( hypothesis - y_data ) )
    w_grad, b_grad = tape.gradient( cost, [W, b] )

    W.assign_sub( learning_rate * w_grad )
    b.assign_sub( learning_rate * b_grad )
    
    # check the variables
    if i%10 == 0:
        print( f"{i:4}  |  W : {W.numpy():^10.4} |  b : {b.numpy():^10.4} |  cost : {cost:^10.6f}" )
    pass
```

## Predict
##### Hypothesis : H(x) = Wx + b

```python
print(W * 5 + b)
print(W * 2.5 + b)
```

## Visualize result

```python
tmp_W = 2.452
tmp_b = 0.376

plt.style.use('ggplot')
plt.plot(x_data, tmp_W * np.array(x_data) + tmp_b, 'r')
plt.plot(x_data, y_data, 'b.')

plt.ylim(0, 8)
plt.title('i=0 (Epoch=1)')
plt.xlabel(f'\nW = {tmp_W}\nb = {tmp_b}', size=13)

plt.show()
```
